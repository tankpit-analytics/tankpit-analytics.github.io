```python
import pandas as pd
```

```python
cup_counts_with_dupes = pd.read_csv('/Users/thomas/Desktop/testrun/cup_counts_with_dupes.csv')
cup_counts = pd.read_csv('/Users/thomas/Desktop/testrun/cup_counts.csv')
```

```python
dupes = cup_counts_with_dupes.groupby(['gold','silver','bronze']).count()['id'].reset_index(drop = False)
dupes = dupes.rename(columns = {'id': 'dupe_count'})
dupes = dupes[dupes['dupe_count'] > 1].reset_index(drop = True)
```

```python
dupes = dupes.merge(cup_counts, on = ['gold','silver','bronze'], how = 'inner').reset_index(drop = True)
```

```python
cup_counts_with_dupes = cup_counts_with_dupes[['id', 'name', 'gold','silver','bronze']]
cup_counts_with_dupes = cup_counts_with_dupes.rename(columns = {'id': 'dupe_id', 'name': 'dupe_name'})
```

```python
dupes = dupes.merge(cup_counts_with_dupes, on = ['gold','silver','bronze'], how = 'inner').reset_index(drop = True)
```
